## Parse the Input Data

```python
def parse(filename):
    """Parse puzzle input data."""
    with open(f'../inputs/{filename}.txt') as f:
        first_line = f.readline().strip()
        len_message = len(first_line)

    cols = [""] * len_message
    with open(f'../inputs/{filename}.txt') as f:
        for line in f:
            for i, c in enumerate(line.strip()):
                cols[i] += c

    return cols
```

```python
parse("day-06_test")
```

## Part 1
---

```python
from collections import Counter
```

```python
def solve(cols):
    message = ""
    for col in cols:
        message += Counter(col).most_common(1)[0][0]
    return message
```

### Run on Test Data

```python
solve(parse("day-06_test")) == "easter"
```

### Run on Input Data

```python
solve(parse("day-06"))
```

## Part 2
---

```python
def solve2(cols):
    message = ""
    for col in cols:
        message += Counter(col).most_common()[-1][0]
    return message
```

### Run on Test Data

```python
solve2(parse("day-06_test")) == "advent"
```

### Run on Input Data

```python
solve2(parse("day-06"))
```

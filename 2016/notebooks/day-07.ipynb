## Parse the Input Data

```python
def parse(filename):
    """Parse puzzle input data."""
    with open(f'../inputs/{filename}.txt') as f:
        lines = [line.strip() for line in f.readlines()]
    return lines
```

```python
import re
```

```python
pattern = re.compile("\[[a-z]+\]")
```

```python
def has_abba(s):
    for i in range(1, len(s)-2):
        if s[i] == s[i+1] and s[i-1] != s[i] and s[i-1] == s[i+2]:
            return True
    return False
```

```python
has_abba('abcdeedfgh') == True
```

```python
def solve(lines):
    valid_count = 0
    for line in lines:
        inside = False
        brackets = re.findall(pattern, line)
        for b in brackets:
            if has_abba(b):
                inside = True
                break

        if inside:
            continue

        outsides = re.split(pattern, line)
        for o in outsides:
            if has_abba(o):
                valid_count += 1
                break

    return valid_count
```

### Run on Test Data

```python
solve(parse("day-07_test")) == 2
```

### Run on Input Data

```python
solve(parse("day-07"))
```

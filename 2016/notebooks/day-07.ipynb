## Parse the Input Data

```python
def parse(filename):
    """Parse puzzle input data."""
    with open(f'../inputs/{filename}.txt') as f:
        lines = [line.strip() for line in f.readlines()]
    return lines
```

## Part 1
---

```python
import re
```

```python
bracket_pattern = re.compile("\[[a-z]+\]")
```

```python
def has_abba(s):
    for i in range(1, len(s)-2):
        if s[i] == s[i+1] and s[i-1] != s[i] and s[i-1] == s[i+2]:
            return True
    return False
```

```python
has_abba('abcdeedfgh') == True
```

```python
def solve(lines):
    valid_count = 0

    for line in lines:
        inside = False
        brackets = re.findall(bracket_pattern, line)
        for b in brackets:
            if has_abba(b):
                inside = True
                break

        if inside:
            continue

        outsides = re.split(bracket_pattern, line)
        for o in outsides:
            if has_abba(o):
                print(o)
                valid_count += 1
                break

    return valid_count
```

### Run on Test Data

```python
solve(parse("day-07_test")) == 2
```

### Run on Input Data

```python
solve(parse("day-07"))
```

## Part 2
---

```python
def get_abas(s):
    abas = []
    for i in range(1, len(s)-1):
        if s[i-1] != s[i] and s[i-1] == s[i+1]:
            abas.append(s[i-1:i+2])
    return abas
```

```python
get_abas("abacyzy")
```

```python
def has_bab(s, aba):
    bab = aba[1] + aba[0] + aba[1]
    if bab in s:
        return True
    return False
```

```python
def solve2(lines):
    valid_count = 0

    for line in lines:
        abas = []
        outsides = re.split(bracket_pattern, line)
        for o in outsides:
            abas.extend(get_abas(o))

        brackets = re.findall(bracket_pattern, line)
        for b in brackets:
            for aba in abas:
                if has_bab(b, aba):
                    valid_count += 1
                    break

    return valid_count
```

### Run on Test Data

```python
solve2(parse("day-07_test2")) == 3
```

### Run on Input Data

```python
solve2(parse("day-07"))
```

## Part 1
---

```python
def decompress(data):
    decompressed = ""

    while True:
        left, right = data.find("("), data.find(")")
        if left == -1:
            decompressed += data[:]
            break
        decompressed += data[:left]
        marker = data[left+1:right]
        chars, n = [int(d) for d in marker.split("x")]
        seq = data[right+1:right+chars+1]
        decompressed += seq * n

        if right + chars + 1 < len(data):
            data = data[right+chars+1:]
        else:
            break

    return decompressed
```

### Run on Test Data

```python
decompress("ADVENT") == "ADVENT"
```

```python
decompress("A(1x5)BC") == "ABBBBBC"
```

```python
decompress("(3x3)XYZ") == "XYZXYZXYZ"
```

```python
decompress("A(2x2)BCD(2x2)EFG") == "ABCBCDEFEFG"
```

```python
decompress("(6x1)(1x3)A") == "(1x3)A"
```

```python
decompress("X(8x2)(3x3)ABCY") == "X(3x3)ABC(3x3)ABCY"
```

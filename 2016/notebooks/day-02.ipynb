## Part 1
---

```python
numpad = {
    (-1, 1): '1',
    (0, 1): '2',
    (1, 1): '3',
    (-1, 0): '4',
    (0, 0): '5',
    (1, 0): '6',
    (-1, -1): '7',
    (0, -1): '8',
    (1, -1): '9'
}
```

```python
deltas = {
    'U': (0, 1),
    'D': (0, -1),
    'L': (-1, 0),
    'R': (1, 0)
}
```

```python
def update_pos(pos, d):
    new_pos = [sum(i) for i in zip(pos, deltas[d])]
    if tuple(new_pos) in numpad:
        return new_pos
    return pos
```

```python
def solve(moves):
    pos = [0, 0]
    nums = []
    for move in moves:
        for d in move:
            pos = update_pos(pos, d)
        nums.append(numpad[tuple(pos)])
    return int("".join(nums))
```

### Run on Test Data

```python
solve(['ULL', 'RRDDD', 'LURDL', 'UUUUD']) == 1985
```

### Run on Input Data

```python
solve(moves)
```

## Part 2
---

```python
keypad = {
   (0, 2): '1',
   (-1, 1): '2',
   (0, 1): '3',
   (1, 1): '4',
   (-2, 0): '5',
   (-1, 0): '6',
   (0, 0): '7',
   (1, 0): '8',
   (2, 0): '9',
   (-1, -1): 'A',
   (0, -1): 'B',
   (1, -1): 'C',
   (0, -2): 'D'
}
```

```python
def update_pos2(pos, d):
    new_pos = [sum(i) for i in zip(pos, deltas[d])]
    if tuple(new_pos) in keypad:
        return new_pos
    return pos
```

```python
def solve2(moves):
    pos = [-2, 0]
    nums = []
    for move in moves:
        for d in move:
            pos = update_pos2(pos, d)
        nums.append(keypad[tuple(pos)])
    return "".join(nums)
```

### Run on Test Data

```python
solve2(['ULL', 'RRDDD', 'LURDL', 'UUUUD']) == '5DB3'
```

### Run on Input Data

```python
solve2(moves)
```

```python
num_pad = {
    (-1, 1): '1',
    (0, 1): '2',
    (1, 1): '3',
    (-1, 0): '4',
    (0, 0): '5',
    (1, 0): '6',
    (-1, -1): '7',
    (0, -1): '8',
    (1, -1): '9'
}
```

```python
deltas = {
    'U': (0, 1),
    'D': (0, -1),
    'L': (-1, 0),
    'R': (1, 0)
}
```

```python
def update_pos(pos, d):
    new_pos = [sum(i) for i in zip(pos, deltas[d])]
    if abs(int(new_pos[0])) > 1 or abs(int(new_pos[1]) > 1):
        return pos
    return new_pos
```

```python
def solve(moves):
    pos = [0, 0]
    nums = []
    for move in moves:
        for d in move:
            pos = update_pos(pos, d)
        nums.append(num_pad[tuple(pos)])
    return "".join(nums)
```

### Run on Test Data

```python
solve(['ULL', 'RRDDD']) #, 'LURDL', 'UUUUD'])
```

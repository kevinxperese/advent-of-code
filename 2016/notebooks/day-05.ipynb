## Part 1
---

```python
import hashlib
```

```python
def get_hash(s):
    m = hashlib.md5()
    m.update(s.encode("utf-8"))
    return m.hexdigest()
```

```python
def solve(door_id):
    pw = ""
    i = 0
    while True:
        hex_hash = get_hash(f"{door_id}{i}")
        if hex_hash.startswith("00000"):
            pw += hex_hash[5]
            if len(pw) == 8:
                return pw
        i += 1
```

### Run on Test Data

```python
get_hash("abc3231929")
```

```python
get_hash("abc5017308")
```

```python
get_hash(f"abc{5278568}")
```

```python
solve('abc') == '18f47a30'
```

### Run on Input Data

```python
solve("ojvtpuvg")
```

## Part 2
---

```python
def solve2(door_id):
    pw = [""] * 8
    i = 0
    while True:
        hex_hash = get_hash(f"{door_id}{i}")
        if hex_hash.startswith("00000"):
            if hex_hash[5].isdigit():
                pos = int(hex_hash[5])
                if pos in range(8) and not pw[pos]:
                    pw[pos] = hex_hash[6]
            if len("".join(pw)) == 8:
                return "".join(pw)
        i += 1
```

### Run on Test Data

```python
solve2('abc') == "05ace8e3"
```

### Run on Input Data

```python
solve2("ojvtpuvg")
```

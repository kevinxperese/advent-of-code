## Parse the Input Data

```python
def parse():
    """Parse puzzle input data."""
    lines = []
    with open(f'../inputs/day-03.txt') as f:
        for line in f.readlines():
            lines.append(tuple(int(n) for n in line.strip().split()))
    return lines
```

```python
lines = parse()
lines[:5]
```

## Part 1
---

```python
def check_triangle(sides):
    c1 = (sides[0] + sides[1] > sides[2])
    c2 = (sides[0] + sides[2] > sides[1])
    c3 = (sides[1] + sides[2] > sides[0])

    return c1 and c2 and c3
```

```python
def solve(lines):
    valid = []
    for line in lines:
        valid.append(check_triangle(line))
    return sum(valid)
```

### Run on Test Data

```python
check_triangle((5, 10, 25))
```

### Run on Input Data

```python
solve(lines)
```

## Part 1
---

```python
import numpy as np
```

```python
def add(grid, x, y):
    grid[:y, :x] = 1
    return grid
```

```python
def shift_right(grid, row, n):
    grid[row] = np.roll(grid[row], n)
    return grid
```

```python
def shift_down(grid, col, n):
    grid[:, col] = np.roll(grid[:, col], n)
    return grid
```

```python
def solve(lines):
    grid = np.zeros((6, 50))

    for line in lines:
        if line.startswith("rect"):
            dims = [int(x) for x in line.split(" ")[1].split("x")]
            grid = add(grid, *dims)
        elif "row" in line:
            args = [int(x) for x in line.split("=")[1].split(" by ")]
            grid = shift_right(grid, *args)
        else:
            args = [int(x) for x in line.split("=")[1].split(" by ")]
            grid = shift_down(grid, *args)

    return int(grid.sum())
```

```python
solve(parse())
```

```python
def solve2(lines):
    grid = np.zeros((6, 50))

    for line in lines:
        if line.startswith("rect"):
            dims = [int(x) for x in line.split(" ")[1].split("x")]
            grid = add(grid, *dims)
        elif "row" in line:
            args = [int(x) for x in line.split("=")[1].split(" by ")]
            grid = shift_right(grid, *args)
        else:
            args = [int(x) for x in line.split("=")[1].split(" by ")]
            grid = shift_down(grid, *args)

    for row in range(6):
        print("".join(["#" if c == 1.0 else " " for c in grid[row]]))
```

```python
solve2(parse())
```

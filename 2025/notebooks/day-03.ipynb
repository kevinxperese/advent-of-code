## Parse the Input Data

```python
def parse(filename):
    """Parse puzzle input data."""
    with open(f'../inputs/{filename}.txt') as f:
        lines = [line.strip() for line in f.readlines()]
    return lines
```

## Part 1
---

```python
def solve(lines):
    total_joltage = 0
    for line in lines:
        tens = max(line[:-1])
        ones = max(line[line.index(tens)+1:])
        total_joltage += int(tens + ones)
    return total_joltage
```

### Run on Test Data

```python
solve(parse("day-03_test")) == 357
```

### Run on Input Data

```python
solve(parse("day-03"))
```

## Parse the Input Data

```python
from collections import defaultdict
```

```python
def parse(filename):
    """Parse puzzle input data."""
    grid = defaultdict(lambda: "#")

    with open(f'../inputs/{filename}.txt') as f:
        for r, row in enumerate(f.readlines()):
            for c, char in enumerate(row.strip()):
                grid[c + r * 1j] = char
                if char == "S":
                    start = c + r * 1j

    return grid, start
```

## Part 1
---

```python
def solve(grid, pos, splitters):
    while True:
        pos += 1j
        val = grid[pos]

        if val == "#":
            break

        if val == "^":
            splitters.add(pos)
            solve(grid, pos-1, splitters)

            if grid[pos+2] != "^":
                solve(grid, pos+1, splitters)

            break

    return len(splitters)
```

### Run on Test Data

```python
solve(*parse("day-07_test"), set()) == 21
```

### Run on Input Data

```python
solve(*parse("day-07"), set())
```

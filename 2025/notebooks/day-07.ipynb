## Parse the Input Data

```python
from collections import defaultdict
```

```python
def parse(filename):
    """Parse puzzle input data."""
    grid = defaultdict(lambda: "#")

    with open(f'../inputs/{filename}.txt') as f:
        for r, row in enumerate(f.readlines()):
            for c, char in enumerate(row.strip()):
                grid[c + r * 1j] = char
                if char == "S":
                    start = c + r * 1j

    return grid, start
```

## Part 1
---

```python
def solve(grid, pos, splitters):
    while True:
        pos += 1j
        val = grid[pos]

        if val == "#":
            break

        if val == "^":
            splitters.add(pos)
            solve(grid, pos-1, splitters)

            if grid[pos+2] != "^":
                solve(grid, pos+1, splitters)

            break

    return len(splitters)
```

### Run on Test Data

```python
solve(*parse("day-07_test"), set()) == 21
```

### Run on Input Data

```python
solve(*parse("day-07"), set())
```

## Part 2
---

```python
def solve2(grid, pos, stop_row):
    num_paths = 0

    while True:
        pos += 1j
        val = grid[pos]

        if pos.imag == stop_row or val == "#":
            num_paths += 1
            break

        if val == "^":
            num_paths += solve2(grid, pos-1, stop_row)
            num_paths += solve2(grid, pos+1, stop_row)
            break

    return num_paths
```

### Run on Test Data

```python
stop_row = max([p.imag for p in parse("day-07_test")[0]])
solve2(*parse("day-07_test"), stop_row) == 40
```

```python
def solve2(filename):
    """Parse puzzle input data."""
    numeric_lines = []
    with open(f'../inputs/{filename}.txt') as f:
        for r, row in enumerate(f.readlines()):
            numeric_lines.append([0] * len(row.strip()))
            for c, char in enumerate(row.strip()):
                if r == 0 and char == "S":
                    numeric_lines[r][c] = 1
                elif char == ".":
                    numeric_lines[r][c] += numeric_lines[r-1][c]
                elif char == "^":
                    numeric_lines[r][c-1] += numeric_lines[r-1][c]
                    numeric_lines[r][c+1] += numeric_lines[r-1][c]

    return sum(numeric_lines[-1])
```

```python
solve2("day-07_test") == 40
```

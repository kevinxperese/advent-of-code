## Parse the Input Data

```python
def parse(filename):
    """Parse puzzle input data."""
    with open(f'../inputs/{filename}.txt') as f:
        reds = [(int(x), int(y)) for line in f for x, y in [line.strip().split(",")]]
    return reds
```

```python
parse("day-09_test")
```

## Part 1
---

```python
from itertools import combinations
```

```python
def solve(reds):
    max_rect_area = 0
    for p1, p2 in combinations(reds, 2):
        x1, y1, x2, y2 = *p1, *p2
        rect_area = (abs(x1 - x2) + 1) * (abs(y1 - y2) + 1)
        max_rect_area = max(max_rect_area, rect_area)
    return max_rect_area
```

### Run on Test Data

```python
solve(parse("day-09_test")) == 50
```

### Run on Input Data

```python
solve(parse("day-09"))
```

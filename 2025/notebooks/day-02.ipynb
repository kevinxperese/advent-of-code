## Parse the Input Data

```python
def parse(filename):
    """Parse puzzle input data."""
    with open(f'../inputs/{filename}.txt') as f:
        ranges = [r for r in f.read().split(",")]
    return ranges
```

## Part 1
---

```python
def solve(ranges):
    invalids = []

    for r in ranges:
        begin, end = [int(x) for x in r.split("-")]

        for i in range(begin, end+1):
            str_i = str(i)
            len_str_i = len(str_i)
            if len_str_i % 2 == 0:
                mid = int(len_str_i / 2)
                if str_i[:mid] == str_i[mid:]:
                    invalids.append(i)

    return sum(invalids)
```

### Run on Test Data

```python
solve(parse("day-02_test")) == 1227775554
```

### Run on Input Data

```python
solve(parse("day-02"))
```

## Part 2
---

```python
import re
```

```python
doubled_numbers = re.compile(r"^(\d+)\1+$")
```

```python
def solve2(ranges):
    invalids = []

    for r in ranges:
        begin, end = [int(x) for x in r.split("-")]

        for i in range(begin, end+1):
            str_i = str(i)
            match = re.match(doubled_numbers, str_i)
            if match:
                    invalids.append(i)

    return sum(invalids)
```

### Run on Test Data

```python
solve2(parse("day-02_test")) == 4174379265
```

### Run on Input Data

```python
solve2(parse("day-02"))
```

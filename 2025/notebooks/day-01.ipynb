## Parse the Input Data

```python
def parse(filename):
    """Parse puzzle input data."""
    with open(f'../inputs/{filename}.txt') as f:
        turns = [line.strip() for line in f.readlines()]
    return turns
```

```python
parse("day-01_test")[:5]
```

## Part 1
---

```python
def solve(turns):
    pos = 50
    count_0 = 0
    d_mult = {'L': -1, 'R': 1}

    for turn in turns:
        d, n = turn[0], int(turn[1:])
        pos = (pos + n * d_mult[d]) % 100
        count_0 += pos == 0

    return count_0
```

### Run on Test Data

```python
solve(parse("day-01_test")) == 3
```

### Run on Input Data

```python
solve(parse("day-01"))
```

## Part 2
---

```python
def solve2(turns):
    pos = 50
    count_0 = 0
    d_mult = {'L': -1, 'R': 1}

    for turn in turns:
        d, n = turn[0], int(turn[1:])

        new_pos = (pos + n * d_mult[d]) % 100
        count_0 += n // 100  # num full rotations

        if d == "L" and new_pos > pos and pos != 0:
            count_0 += 1
        elif d == "R" and new_pos < pos and new_pos != 0:
            count_0 += 1

        count_0 += (new_pos == 0)

        pos = new_pos

    return count_0
```

### Run on Test Data

```python
solve2(parse("day-01_test")) == 6
```

### Run on Input Data

```python
solve2(parse("day-01"))
```
